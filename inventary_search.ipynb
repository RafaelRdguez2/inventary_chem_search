```python
# @title Importar las librerías necesarias
! pip install fastprogress
from fastprogress import master_bar, progress_bar

#Importa las librerías necesarias
!pip install rdkit
import rdkit
import pandas as pd
from os import name
import io

from IPython.display import clear_output
import time
```

```python
# @title Toque para subir el fichero excel del inventario
#Permite subir el inventario

from google.colab import files
uploaded = files.upload()

for fn in uploaded.keys():
  name = fn
  print('Se ha subido el archivo exitosamente el archivo: "{name}" que ocupa: {length} bytes'.format(
      name=name, length=len(uploaded[fn])))


df = pd.read_excel(io.BytesIO(uploaded[name]))
```

```python
# @title Ingrese la substructura a buscar en formato SMILES {"display-mode":"form"}
q = "C1CCCC1" # @param {"type":"string","placeholder":"SMILE"}
```

```python
# @title Realiza la búsqueda de todas las estructuras en el invetario
mol_q = rdkit.Chem.MolFromSmiles(q)
molecules = df[["SMILES"]].values.flatten().tolist()

from IPython.display import display
mols = []
for i, mol in enumerate(molecules):
  try:
    mol_m = rdkit.Chem.MolFromSmiles(str(mol))
    if mol_m.HasSubstructMatch(mol_q):
      mols.append(df.loc[i])
  except:
    pass

time.sleep(1) # Wait a moment
clear_output(wait=True) # Clear the output

df_mols = pd.DataFrame(mols)
display(df_mols)
```
